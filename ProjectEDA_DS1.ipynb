```python
import pandas as pd

# Load the dataset from the provided link
df_sales = pd.read_csv(r"C:\Users\Hello\Downloads\archive (2).zip", parse_dates=["Date"], index_col="Date")

# Rename columns to more unique names
column_rename_map = {
    'Transaction ID': 'TransID',
    'Customer ID': 'CustID',
    'Gender': 'CustGender',
    'Age': 'CustAge',
    'Product Category': 'ProdCategory',
    'Quantity': 'QtySold',
    'Price per Unit': 'UnitPrice',
    'Total Amount': 'TotalSales',
    'Date': 'SaleDate'
}

df_sales.rename(columns=column_rename_map, inplace=True)
```

```python
# Calculate mean
mean_qty = df_sales['QtySold'].mean()
mean_price = df_sales['UnitPrice'].mean()
mean_total = df_sales['TotalSales'].mean()

print(f"The mean of QtySold is: {mean_qty}")
print(f"The mean of UnitPrice is: {mean_price}")
print(f"The mean of TotalSales is: {mean_total}\n")

# Calculate median
median_qty = df_sales['QtySold'].median()
median_price = df_sales['UnitPrice'].median()
median_total = df_sales['TotalSales'].median()

print(f"The median of QtySold is: {median_qty}")
print(f"The median of UnitPrice is: {median_price}")
print(f"The median of TotalSales is: {median_total}\n")

# Calculate mode
mode_qty = df_sales['QtySold'].mode().iloc[0]
mode_price = df_sales['UnitPrice'].mode().iloc[0]
mode_total = df_sales['TotalSales'].mode().iloc[0]

print(f"The mode of QtySold is: {mode_qty}")
print(f"The mode of UnitPrice is: {mode_price}")
print(f"The mode of TotalSales is: {mode_total}\n")

# Calculate standard deviation
std_qty = df_sales['QtySold'].std()
std_price = df_sales['UnitPrice'].std()
std_total = df_sales['TotalSales'].std()

print(f"The standard deviation of QtySold is: {std_qty}")
print(f"The standard deviation of UnitPrice is: {std_price}")
print(f"The standard deviation of TotalSales is: {std_total}\n")
```

```python
import matplotlib.pyplot as plt

# Resample the data to monthly frequency and sum the TotalSales
monthly_sales = df_sales.resample('ME')['TotalSales'].sum()

# Plotting the time series
plt.figure(figsize=(12, 6))
plt.plot(monthly_sales, marker='o', linestyle='-')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.grid(True)
plt.show()
```

```python
# Customer Gender Distribution
gender_dist = df_sales['CustGender'].value_counts()

plt.figure(figsize=(8, 5))
plt.bar(gender_dist.index, gender_dist.values, color='teal', edgecolor='black')
plt.title('Customer Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()

# Customer Age Distribution
plt.figure(figsize=(10, 6))
plt.hist(df_sales['CustAge'], bins=20, color='coral', edgecolor='black')
plt.title('Customer Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
import seaborn as sns

# Product Category Analysis
category_monthly_sales = df_sales.pivot_table(index=df_sales.index.month, columns='ProdCategory', values='TotalSales', aggfunc='sum')

plt.figure(figsize=(15, 10))
sns.heatmap(category_monthly_sales, cmap='coolwarm', annot=True, fmt='.0f', linewidths=.5)
plt.title('Monthly Product Category Sales Heatmap')
plt.xlabel('Product Category')
plt.ylabel('Month')
plt.xticks(rotation=45)
plt.show()
```
